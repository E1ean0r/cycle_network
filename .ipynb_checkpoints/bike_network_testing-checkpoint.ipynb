```python
df = pd.DataFrame(
    {'City': ['Aberdeen'],
     'Latitude': [57.634446881002624],
     'Longitude': [-2.247975102064132]})
origin = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
origin = origin.set_crs(epsg=4326)
origin = origin.to_crs(epsg=3857)

origin.head()
```

```python
### calculate closest ditance 
df_n = gpd.sjoin_nearest(origin, paths).merge(paths, left_on="index_right", right_index=True)
df_n["distance"] = df_n.apply(lambda r: r["geometry_x"].distance(r["geometry_y"]), axis=1)

df_n
```

```python
graph = ox.graph_from_place('Scotland, United Kingdom', network_type='bike')
```

```python
df_n = df_n.rename(columns={'geometry_y': 'geometry'})

# Create a new GeoDataFrame destination from df_n
destination = gpd.GeoDataFrame(df_n[['RoadClass_y', 'SHAPE_Leng_y', 'geometry', 'distance']], geometry='geometry')
destination.head()
```

```python
# Get the nearest network nodes to the origin point
nearest_origin = ox.distance.nearest_nodes(graph, X=origin.geometry.x, Y=origin.geometry.y, return_dist=False)
```

```python
# Get the nearest network nodes to the destination linestring
nearest_destination = ox.distance.nearest_nodes(graph, X=[coord[0] for coord in destination.geometry.iloc[0].coords], Y=[coord[1] for coord in destination.geometry.iloc[0].coords], return_dist=False)
```

```python
# Create a subgraph containing only the nearest nodes
nearest_nodes = list(set(nearest_origin) | set(nearest_destination))
subgraph = nx.subgraph(graph, nearest_nodes)
```

```python
# Get the shortest path from the origin to the destination
route = nx.shortest_path(subgraph, nearest_origin[0], nearest_destination[0], weight='length')
```

```python
# Create a GeoDataFrame for the route
route_nodes = ox.graph_to_gdfs(subgraph, edges=False)
route_line = LineString([(route_nodes.loc[node, 'x'], route_nodes.loc[node, 'y']) for node in route])
route_gdf = gpd.GeoDataFrame({'geometry': [route_line], 'route': [route]}, crs=origin.crs)

route_gdf
```

```python

# Iterate over the linestrings
for index, row in paths.iterrows():
    linestring = row['geometry']
    # Convert linestring to points
    points = [Point(coord) for coord in linestring.coords]
    # Find the nearest network nodes to the linestring points
    nearest_nodes = ox.distance.nearest_nodes(graph, X=points, Y=points, return_dist=False)
    # Create a subgraph containing only the nearest nodes
    subgraph = nx.subgraph(graph, nearest_nodes)
    # Find the nearest network nodes to the starting and ending points
    start_nearest_node = ox.distance.nearest_nodes(graph, X=points[0], return_dist=False)
    end_nearest_node = ox.distance.nearest_nodes(graph, X=points[-1], return_dist=False)
    # Calculate the shortest path using Dijkstra's algorithm
    route = nx.shortest_path(subgraph, source=start_nearest_node, target=end_nearest_node, weight='length')
    # Convert the route to LineString
    route_line = LineString([(graph.nodes[node]['x'], graph.nodes[node]['y']) for node in route])
    # Calculate the total length of the route in meters
    total_length = sum(geodesic((graph.nodes[u]['y'], graph.nodes[u]['x']),
                                (graph.nodes[v]['y'], graph.nodes[v]['x'])).meters for u, v in zip(route[:-1], route[1:]))
    # Append the result to the shortest_routes_df DataFrame
    shortest_routes_df = shortest_routes_df.append({'geometry': route_line, 'route': route, 'total_length': total_length}, ignore_index=True)

# Print the shortest routes DataFrame
print(shortest_routes)
```
